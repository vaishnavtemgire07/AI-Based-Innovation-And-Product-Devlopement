```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Downloads/diabetes.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.drop(columns=['Pregnancies','SkinThickness','Insulin','DiabetesPedigreeFunction'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=['Outcome'])
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
le=LogisticRegression()
```

```python
le.fit(x_train,y_train)
```

```python
y_pred=le.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
accuracy_le=accuracy_score(y_test,y_pred)
cr=classification_report(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
```

```python
accuracy_le
```

```python
print(cr)
```

```python
cm
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred1=model.predict(x_test)
y_pred1
```

```python
accuracy_svm=accuracy_score(y_test,y_pred1)
cr=classification_report(y_test,y_pred1)
cm=confusion_matrix(y_test,y_pred1)
```

```python
accuracy_svm
```

```python
print(cr)
```

```python
cm
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
models=['LogisticRegression','SVM']
accuracies=[accuracy_le,accuracy_svm]
plt.figure(figsize=(10,8))
plt.bar(models,accuracies,color=['Red','Blue'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.show()
```
